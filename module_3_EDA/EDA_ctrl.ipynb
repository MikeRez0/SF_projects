```python
def iqr_plot(data):
    median = data.median()
    IQR = data.quantile(0.75) - data.quantile(0.25)
    perc25 = data.quantile(0.25)
    perc75 = data.quantile(0.75)
    print('25-й перцентиль: {},'.format(perc25), '75-й перцентиль: {},'.format(perc75), "IQR: {}, ".format(IQR),"Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5*IQR, l=perc75 + 1.5*IQR))
    data.hist(label = 'data', bins=len(data.value_counts()))
    data.loc[data.between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)].hist(alpha= 0.5, label = 'IQR')
    plt.legend()
    plt.show()
```

```python
#Диагностика значений колонки:
# - NA, уникальных значений
# - Распределение значенний (iqr и/или гистограмма)
def diag_col(colname, iqr=False, value_counts=False):
    print('----------------------')
    print(f'Первичная диагностика {Back.GREEN} {colname} {Style.RESET_ALL}')
    print(f'Описание: {coldesc[colname]}')
    print(f'NA значений: {stud[colname].isna().sum()/len(stud[colname]):.2}')
    print(f"Уникальных значений: {stud[colname].nunique()}")

    print('Справка по значениям: ')
    vc = stud[colname].value_counts(dropna=False, normalize=True)
    if iqr:
        iqr_plot(stud[colname])
        display(stud[colname].describe())        
    if value_counts:
        vc.plot.bar()
        plt.show()   
```

## Тесты Стьюдента

```python
def get_stat_dif(column):
    ycol = 'score'
    cols = stud.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud.loc[stud.loc[:, column] == comb[0], ycol], 
                        stud.loc[stud.loc[:, column] == comb[1], ycol]).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
# Выбираем категориальные и ординальные признаки
for col in cc:
    get_stat_dif(col)
```
